```python
from sklearn.datasets import fetch_openml

X, y = fetch_openml('UMIST_Faces_Cropped', version=1, return_X_y=True)
```

```python
import numpy as np

data = []
for _y in np.unique(y):
    data.append(X[np.where(y==_y)])
```

```python
import matplotlib.pyplot as plt


for x in data[0]:
    plt.imshow(x.reshape(112, 92), cmap=plt.get_cmap('gray'))
    plt.show()
```
